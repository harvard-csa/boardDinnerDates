```python
import random
import scipy.stats as ss
import math
```

```python
#Add senior council?

BOARD = ['Danika', 'Hannah', 'Amy', 'Chris',
        'Kathy', 'Carol', 'Alvin', 'Will', 
        'David', 'Tex', 'Kalos', 'Glen',
        'Shwe', 'Vick', 'Margo', 'Audrey G', 
        'Eric', 'Janice', 'Daniel', 'Cody', 
        'Christy', 'Steph', 'Audrey K', 'Christina',
        'Ricky', 'Serena', 'Sophia']
GROUP_SIZE = 4
boardSize = len(BOARD)
EVEN_OUT_INDEX = 24
# list of dicts
history = []
for _ in range(boardSize):
    history.append({'Danika':0, 'Hannah':0, 'Amy':0, 'Chris':0,
        'Kathy':0, 'Carol':0, 'Alvin':0, 'Will':0, 
        'David':0, 'Tex':0, 'Kalos':0, 'Glen':0,
        'Shwe':0, 'Vick':0, 'Margo':0, 'Audrey G':0, 
        'Eric':0, 'Janice':0, 'Daniel':0, 'Cody':0, 
        'Christy':0, 'Steph':0, 'Audrey K':0, 'Christina':0,
        'Ricky':0, 'Serena':0, 'Sophia':0})
history[BOARD.index('Chris')]['Christina'] += 2
history[BOARD.index('Christina')]['Chris'] += 2
```

```python
def initialize():
    # random numbers for everyone
    randomNums = [random.random() for _ in range(boardSize)]
    
    # converts random floats to rank ordering
    ranks = ss.rankdata(randomNums)
    
    # matches a random number 1-30 to board members
    board = zip(BOARD, ranks)
    
    # sort members by random number (1-4 is a group, 2-8...)
    board = sorted(board, key = lambda p: p[1])
    
    
    # check if redo is necessary (no members meet twice)
    for i in range(0, boardSize, 4):
        # go through first three members in the group
        for j in range(i, i+3):
            # check each member's history
            for k in range(i, i+3):
                if j != k and history[BOARD.index(board[j][0])][board[k][0]] > 1:
                    return('redo')
        # go through fourth member's history in groups of 4
        if i != EVEN_OUT_INDEX:
            for k in range(i, i+3):
                 if history[BOARD.index(board[i+3][0])][board[k][0]] > 1:
                    return('redo')    
      

    # add to member history and print out groups
    for i in range(0, boardSize, 4):
        for j in range(i, i+3):
            for k in range(i, i+3):
                if j != k:
                    history[BOARD.index(board[j][0])][board[k][0]] += 1

        # store group
        tmp = [board[i][0], board[i+1][0], board[i+2][0]]
        
        # deal with fourth member
        if i != EVEN_OUT_INDEX:
            for k in range(i, i+3):
                 history[BOARD.index(board[i+3][0])][board[k][0]] += 1
            tmp += [board[i+3][0]]
        
        # print gruops
        (print(', '.join(tmp)))
        

if __name__ == "__main__":
    for i in range(1,17):
        print("Week", i,": ")
        while initialize() == 'redo':
            pass
```
